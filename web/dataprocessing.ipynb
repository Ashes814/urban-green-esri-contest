```python
import pandas as pd
import plotly.graph_objs as go
import datetime
```

```python
df = pd.read_csv('Covid_data.csv', encoding='utf-8')
df.head()
```

```python
df = df[df['provinceName']=='上海市']
```

```python
useless_fields = ['continentName','continentEnglishName','countryName',
                  'countryEnglishName', 'province_zipCode']
df = df.drop(useless_fields, axis=1)
```

```python
import mitosheet
mitosheet.sheet(df, view_df=True)
```

```python
df.info()
```

```python
df['updateTime'] = pd.to_datetime(df['updateTime'],
                                  format = '%Y-%m-%d')
```

```python
df['date'] = df['updateTime'].dt.strftime('%Y-%m-%d')
```

```python
df_out = df[df['cityName'] == '境外输入']
df_in = df[df['cityName'] != '境外输入']
```

```python
df_out_sh = df_out.groupby(['date']).sum().reset_index()
df_out_sh = df_out_sh[['date', 'city_confirmedCount', 'city_suspectedCount',
                       'city_curedCount', 'city_deadCount']]

df_in_sh = df_in.groupby(['date']).sum().reset_index()
df_in_sh = df_in_sh[['date', 'city_confirmedCount', 'city_suspectedCount',
                       'city_curedCount', 'city_deadCount']]
```
